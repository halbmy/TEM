# Using the TEM class for data inversion
## Central-loop soundings with sTEM

We make use of the data published by Madsen et al. (2026), made publicly available on [Zenodo](https://doi.org/10.5281/zenodo.16894039). Here we focus on the central-loop data with 40x40 and 80x80m loops.

We import the main class `TEM`.

```python
from tem import TEM
```

The instance is instantiated through the data file name. Instead of the USF files, we chose the xyz data files as it includes all soundings in one go. Whereas the data are in the xyz file, a gex file is exported containing the geometry of the loops, the time gates and the current waveform. If a file with the same name like the xyz file but the ending gex is found, it is read by default. Alternatively, you can specify this file explicitly by `cfg=`.

```python
self = TEM("data/madsen2026/40x40.xyz") # , cfg="data/madsen2026/40x40.gex")
print(self)
```

In the file, there are also the GPS positions that can be shown:

```python
self.showPositions();
```

A single sounding can be shown by `showSounding()` with the given number, by default as transient. By specifying `rhoa=True`, you see a quick-and-dirty apparent resistivity curve.

```python
self.showSounding(0);
```

You can also plot several soundings by specifying their numbers (by default, all). Again, `rhoa` can be used.

```python
self.showSoundings(rhoa=True);
```

A single sounding can be inverted by `invertSounding()`. By default (`show=True`), model and data fit are plotted (which can also be done explicitly by `show1dResult()`).

```python
self.invertSounding(0)
```

We now load data from the 80x80m loop, here only three (locations not shown).

```python
self = TEM("data/madsen2026/80x80.xyz")
print(self)
```

```python
self.showSoundings(rhoa=True);
```

```python
self.invertSounding()
```

At any time, the current waveform can be shown.

```python
ax = self.showWaveform();
```

We see mainly the repetion rate. Like all the `show*` functions, the axis is returned and can be used for refining the plot.

```python
ax = self.showWaveform();
ax.set_xlim(0, 3e-5)
ax.set_ylim(0, 1)
```
