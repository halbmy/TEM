# Using the TEM class for data inversion
## Central-loop soundings with sTEM

We make use of the data published by Madsen et al. (2026), made publicly available on [Zenodo](https://doi.org/10.5281/zenodo.16894039). Here we focus on the central-loop data with 40x40 and 80x80m loops.

We import the main class `TEM`.

```python
from tem import TEM
```

The instance is instantiated through the data file name. Instead of the USF files, we chose the xyz data files as it includes all soundings in one go. Whereas the data are in the xyz file, a gex file is exported containing the geometry of the loops, the time gates and the current waveform. If a file with the same name like the xyz file but the ending gex is found, it is read by default. Alternatively, you can specify this file explicitly by `cfg=`.

```python
tem40 = TEM("data/madsen2026/40x40.xyz") # , cfg="data/madsen2026/40x40.gex")
print(tem40)
```

```python
tem80 = TEM("data/madsen2026/80x80.xyz") # , cfg="data/madsen2026/40x40.gex")
print(tem80)
```

```python
ax = tem40.showWaveform();
```

```python
ax = tem80.showWaveform();
```

```python
ax = tem40.showWaveform(label="40x40")
tem80.showWaveform(ax=ax, label="80x80")
ax.set_xlim(-0e-6, 20e-6)
ax.set_ylim(0, 1)
```
